```python
# 程式碼A區==程式碼A區==程式碼A區==程式碼A區==程式碼A區
# AAAAAAAA==AAAAAAAA==AAAAAAAAA==AAAAAAAA==AAAAAAAA
#
# Dodecagon-2.ipynb
# Beauty of Islamic Pattern
# 2021-09-16
#
# python 3.8
# run on https://replit.com
# run on ubuntu 16.04 LTS
#
# 要執行本程式, 請按上方run▶
#
# Browser encoding UTF-8 HTML5 URL / as %2F
#
# 【觀念】sin、cos、tan
# https://www.youtube.com/watch?v=8o_rBszqkOg
#
# 程式碼B區==程式碼B區==程式碼B區==程式碼B區==程式碼B區
# BBBBBBBB==BBBBBBBB==BBBBBBBBB==BBBBBBBB==BBBBBBBB
#
# to fix matplotlib warning
import os
os.environ['MPLCONFIGDIR'] = os.getcwd() + "/configs/"
#
import sys
#
# 載入繪圖模組 pyplot, 重新命名為 plot
import matplotlib.pyplot as plot
import numpy as np
from shapely  import affinity
from shapely.geometry  import LineString, Point, Polygon
#
# 定義圖面函數 ------------------------------------------------------
def set_graphic_area(width,height) :
    #    
    cm2inch = 1/2.54    # inch per cm
    #
    # define graphic area
    #
    left_margin = 1.0   # cm
    right_margin = 1.0  # cm
    #
    figure_width  = width  # cm , from xmin to xmax
    figure_height = height # cm , from ymin to ymax
    #
    top_margin = 1.0    # cm
    bottom_margin = 1.0 # cm
    #
    box_width = left_margin + figure_width + right_margin   # cm
    box_height = top_margin + figure_height + bottom_margin # cm
    #
    top_value    = 1.0 - top_margin / box_height
    bottom_value = bottom_margin / box_height
    left_value   = left_margin / box_width
    right_value  = 1.0 - right_margin / box_width
    #
    return (box_width*cm2inch,box_height*cm2inch,top_value,bottom_value,left_value,right_value,width)
#
# end of def
#
# Define the arc
# center position is cxy = (cx, cy)
# start_angle, end_angle is in degrees
#
def shapely_Arc(cxy, r, start_angle, end_angle) :
    #
    numsegments = 1440
    # The coordinates of the arc
    theta = np.radians(np.linspace(start_angle, end_angle, numsegments))
    x = cxy[0] + r * np.cos(theta)
    y = cxy[1] + r * np.sin(theta)
    Arc = LineString(np.column_stack([x, y]))
    return Arc
    #
# end of def
#
def motif_line(LS, COLOR, ZORDER, LINEWIDTH) :
    #
    xs, ys = LS.xy
    ax.plot(xs, ys, color=COLOR, alpha=0.7, linewidth=LINEWIDTH, solid_capstyle='round', zorder=ZORDER)
    return
#
def motif_fill(LS, COLOR, ZORDER, ALPHA) :
    #
    xs, ys = LS.xy
    ax.fill(xs, ys, alpha=ALPHA, closed=True, edgecolor=None, facecolor=COLOR, fill=True, joinstyle='miter', zorder=ZORDER)
    return
#
def polygon_outline(SHAPE, MYcolor, ZORDER, LINEWIDTH) :
    #
    xs, ys = SHAPE.exterior.coords.xy
    ax.plot(xs, ys, color=MYcolor, alpha=1.0, linewidth=LINEWIDTH, solid_capstyle='round', zorder=ZORDER)
    return
#
def polygon_fill(SHAPE, color, ZORDER, Alpha) :
    #
    xs, ys = SHAPE.exterior.coords.xy
    ax.fill(xs, ys, alpha=Alpha, closed=True, edgecolor=None, facecolor=color, fill=True, zorder=ZORDER)
    return
#
# 程式碼C區==程式碼C區==程式碼C區==程式碼C區==程式碼C區
# CCCCCCCC==CCCCCCCC==CCCCCCCCC==CCCCCCCC==CCCCCCCC
#
# 設定圖面大小
#  setup圖面大小 (10,10) 單位公分
#  tup7 is type of Tuple, have 7 indexed value
#
# main()
tup7 = set_graphic_area(20,20)  
# 呼叫 pyplot.figure(), 建立一個圖表物件, 並成為目前圖表物件
fig = plot.figure(figsize=(tup7[0], tup7[1])) 
# 圖表的繪圖區域被分為1個子圖, 1 row, 1 column, plot to index 1
ax = fig.add_subplot(1,1,1)   
fig.subplots_adjust(
                top    = tup7[2] ,
                bottom = tup7[3] ,
                left   = tup7[4] ,
                right  = tup7[5] ,
                )
#
plot.xlim(-tup7[6]/2, tup7[6]/2)    # 設定X軸的顯示範圍, from xmin to xmax
plot.ylim(-tup7[6]/2, tup7[6]/2)    # 設定Y軸的顯示範圍, from ymin to ymax
#
xvalues = [-10,-5,0,5,10]
yvalues = [-10,-5,0,5,10]
plot.xticks(xvalues)
plot.yticks(yvalues)
plot.tick_params(labelsize=10)
#
# get current axes, set X,Y same ratio & scale 
plot.gca().set_aspect('equal', adjustable='box')
#
plot.grid(color='green', linewidth=0.3)
ax.grid(True)
# ax.grid(False)
ax.set_axis_on() # will turn on  grid line
# ax.set_axis_off()  # will turn off grid line
#
plot.savefig("./png/2-0.png", format="png", dpi=150)
#
# 程式碼D區==程式碼D區==程式碼D區==程式碼D區==程式碼D區
# DDDDDDDD==DDDDDDDD==DDDDDDDDD==DDDDDDDD==DDDDDDDD
#
r0 = 5
cir_0 = shapely_Arc((0,5), r0, 0, 360)
motif_line(cir_0, 'green', 10, 0.8)
#
plot.savefig("./png/2-1A.png", format="png", dpi=150)
#
xy0 = (r0*np.cos( 0*np.pi/6), 5+r0*np.sin( 0*np.pi/6))
xy1 = (r0*np.cos( 1*np.pi/6), 5+r0*np.sin( 1*np.pi/6))
xy2 = (r0*np.cos( 2*np.pi/6), 5+r0*np.sin( 2*np.pi/6))
xy3 = (r0*np.cos( 3*np.pi/6), 5+r0*np.sin( 3*np.pi/6))
xy4 = (r0*np.cos( 4*np.pi/6), 5+r0*np.sin( 4*np.pi/6))
xy5 = (r0*np.cos( 5*np.pi/6), 5+r0*np.sin( 5*np.pi/6))
xy6 = (r0*np.cos( 6*np.pi/6), 5+r0*np.sin( 6*np.pi/6))
xy7 = (r0*np.cos( 7*np.pi/6), 5+r0*np.sin( 7*np.pi/6))
xy8 = (r0*np.cos( 8*np.pi/6), 5+r0*np.sin( 8*np.pi/6))
xy9 = (r0*np.cos( 9*np.pi/6), 5+r0*np.sin( 9*np.pi/6))
xyA = (r0*np.cos(10*np.pi/6), 5+r0*np.sin(10*np.pi/6))
xyB = (r0*np.cos(11*np.pi/6), 5+r0*np.sin(11*np.pi/6))
#
Dodecagon_xy = [xy0, xy1, xy2, xy3, xy4, 
                xy5, xy6, xy7, xy8, xy9, 
                xyA, xyB, xy0]
Dodecagon_1 = LineString(Dodecagon_xy)
Dodecagon_2 = affinity.rotate(Dodecagon_1, 120, (0,0))
Dodecagon_3 = affinity.rotate(Dodecagon_1, 240, (0,0))
#
motif_fill(Dodecagon_1, 'yellow', 10, 0.1)
motif_line(Dodecagon_1, 'black',  20, 0.6)
#
plot.savefig("./png/2-1B.png", format="png", dpi=150)
#
motif_fill(Dodecagon_2, 'yellow', 10, 0.1)
motif_fill(Dodecagon_3, 'yellow', 10, 0.1)
motif_line(Dodecagon_2, 'black',  20, 0.6)
motif_line(Dodecagon_3, 'black',  20, 0.6)
#
plot.savefig("./png/2-2.png", format="png", dpi=150)
#
edge_7_9_xy = [xy7, xy9]
edge_789_xy = [xy7, xy8, xy9]
edge_7_9 = LineString(edge_7_9_xy)
edge_789 = LineString(edge_789_xy)
motif_line(edge_789, 'blue', 20, 0.8)
#
plot.savefig("./png/2-3A.png", format="png", dpi=150)
#
mid_1 = edge_7_9.interpolate(0.5, normalized = True)
mid_1_xy = (mid_1.x, mid_1.y)
r180_edge_789 = affinity.rotate(edge_789, 180, mid_1_xy)
motif_line(r180_edge_789, 'blue', 20, 0.8)
#
plot.savefig("./png/2-3B.png", format="png", dpi=150)
#
rhombus_1_xy = [xy7, xy8, xy9, r180_edge_789.coords[1], r180_edge_789.coords[2]]
rhombus_1    = LineString(rhombus_1_xy)
rhombus_2 = affinity.rotate(rhombus_1, 120, (0,0))
rhombus_3 = affinity.rotate(rhombus_1, 240, (0,0))
motif_fill(rhombus_1, 'violet', 30, 0.5)
#
plot.savefig("./png/2-3C.png", format="png", dpi=150)
#
motif_fill(rhombus_2, 'violet', 30, 0.5)
motif_fill(rhombus_3, 'violet', 30, 0.5)
#
plot.savefig("./png/2-3D.png", format="png", dpi=150)
#
p00 = Point(0,0)
p00_xy = (0,0)
p01_xy = r180_edge_789.coords[1]
p02 = affinity.rotate(p00, 90, p01_xy)
p02_xy = (p02.x, p02.y)
p03_xy = (-p01_xy[0], p01_xy[1])
#
square_xy =[p00_xy, p01_xy, p02_xy, p03_xy, p00_xy]
square_1  = LineString(square_xy)
square_2  = affinity.rotate(square_1, 120, (0,0))
square_3  = affinity.rotate(square_1, 240, (0,0))
motif_fill(square_1, 'blue', 30, 0.3)
motif_line(square_1, 'green', 30, 0.5)
#
plot.savefig("./png/2-4.png", format="png", dpi=150)
#
posi_1_xy = (0,5)
posi_1 = Point(posi_1_xy)
posi_2 = affinity.rotate(posi_1, 120, (0,0))
posi_3 = affinity.rotate(posi_1, 240, (0,0))
posi_2_xy = (posi_2.x, posi_2.y)
posi_3_xy = (posi_3.x, posi_3.y)
#
for i in (1,2,3,4,5) :
    r_square = affinity.rotate(square_1, i*60, posi_1_xy)
    motif_fill(r_square, 'blue',  30, 0.3)
    motif_line(r_square, 'green', 30, 0.5)
# end for
#
plot.savefig("./png/2-5.png", format="png", dpi=150)
#
for i in (0,1,2,3,4,5) :
    r_square = affinity.rotate(square_2, i*60, posi_2_xy)
    motif_fill(r_square, 'blue',  30, 0.3)
    motif_line(r_square, 'green', 30, 0.5)
# end for
#
plot.savefig("./png/2-6.png", format="png", dpi=150)
#
for i in (0,1,2,3,4,5) :
    r_square = affinity.rotate(square_3, i*60, posi_3_xy)
    motif_fill(r_square, 'blue',  30, 0.3)
    motif_line(r_square, 'green', 30, 0.5)
# end for
#
plot.savefig("./png/2-7.png", format="png", dpi=150)
#
```
